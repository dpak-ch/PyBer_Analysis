```python
# Import dependencies
import matplotlib.pyplot as plt
import statistics
import numpy as np
```

```python
# %matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis,y_axis)
```

```python
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("Fare by Month")
plt.legend()
plt.grid()
plt.figure(figsize=(20,10))
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis,color="green",marker="D",linewidth=2,label='Boston')
ax.grid()
ax.set_ylim(0,45)
ax.set_xlabel("Date")
ax.set_ylabel("Fare")
```

```python
plt.bar(x_axis,y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.bar(x_axis,y_axis,color="black",label="New York")
ax.grid()
ax.set_xlabel("Date")
ax.set_ylabel("Fare")
ax.set_title("Pyber Results")
```

```python
# plt.plot(x_axis,y_axis,"o")
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3) 
plt.scatter(x_axis,y_axis,color="red",label="Chicago",s=y_axis_larger)
plt.legend()
plt.ylim(0,50)
plt.xlabel("Date")
plt.ylabel("Fare")
plt.title("PyBer Fare by Month")
```

```python
fig = plt.figure()
ax = plt.subplot()
ax.scatter(y_axis,x_axis,s=[y*4 for y in y_axis],color='skyblue',alpha=0.8,edgecolors='black',linewidth=2,label='Boston')
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare')
plt.gca().invert_yaxis()
ax.set_xlim(0,50)
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(12, 12))
plt.pie(y_axis,labels=x_axis,explode=explode_values,autopct='%.1f%%')
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.yticks(np.arange(0, 51, step=5.0))
```
